🟢 SECTION 2: Attrition Segmentation

GENDER_WISE_ATTRITION

```python
male_attrition=(ibm_emp_att[(ibm_emp_att["Attrition"]=="Yes") & (ibm_emp_att["Gender"]=="Male")]["EmployeeNumber"].count()/ibm_emp_att[ibm_emp_att["Gender"]=="Male"]["EmployeeNumber"].count())*100
female_attrition=(ibm_emp_att[(ibm_emp_att["Attrition"]=="Yes") & (ibm_emp_att["Gender"]=="Female")]["EmployeeNumber"].count()/ibm_emp_att[ibm_emp_att["Gender"]=="Female"]["EmployeeNumber"].count())*100
print(f"Male_Attrition: {round(male_attrition,2).astype(str)+"%"}\nFemale_Attrition: {round(female_attrition,2).astype(str)+"%"}")
"""Male employees show a slightly higher attrition rate (17%) 
compared to female employees (14.8%),
indicating a mild gender-based difference, though not significantly large."""
```

DEPARTMENT WISE ATTRITIONS

```python
dept=[]
attrition_rate=[]
for i in ibm_emp_att["Department"].unique():
    total=ibm_emp_att[ibm_emp_att["Department"]==i]["EmployeeNumber"].count()
    attrition=ibm_emp_att[(ibm_emp_att["Attrition"]=="Yes")&(ibm_emp_att["Department"]==i)]["EmployeeNumber"].count()
    dept_att=(attrition/total)*100
    dept.append(i)
    attrition_rate.append(dept_att)
    print(f"{i}:{round(dept_att,2)}%")

plt.figure(figsize=(10,6))
ax=sns.barplot(x=dept,y=attrition_rate,hue=dept,palette=["yellow","blue","orange"],legend=True,width=0.5)
plt.title("DEPARTMENT WISE ATTRITIONS",color="red")
plt.ylabel("Attrition %",color="red")
plt.xlabel("Department",color="red")
for container in ax.containers:
    ax.bar_label(container,fmt="%.2f%%")
plt.show()
```

Age Group vs Attrition Rate

```python
Age_Group=[]
Attrition=[]
Early_Career=ibm_emp_att[(ibm_emp_att["Age"]>=18) & (ibm_emp_att["Age"]<=25)]["EmployeeNumber"]
Developing_Career=ibm_emp_att[(ibm_emp_att["Age"]>25) & (ibm_emp_att["Age"]<=35)]["EmployeeNumber"]
Mid_Career=ibm_emp_att[(ibm_emp_att["Age"]>35) & (ibm_emp_att["Age"]<=45)]["EmployeeNumber"]
Senior=ibm_emp_att[(ibm_emp_att["Age"]>45)]["EmployeeNumber"]
groups={"Early_Career":Early_Career,
       "Developing_Career":Developing_Career,
       "Mid_Career":Mid_Career,
       "Senior":Senior}
for name,group in groups.items():
    total=group.count()
    attrition=ibm_emp_att[(ibm_emp_att["Attrition"]=="Yes")&(ibm_emp_att["EmployeeNumber"].isin(group))]["EmployeeNumber"].count()
    Age_Group_Atrr=(attrition/total)*100
    Age_Group.append(name)
    Attrition.append(Age_Group_Atrr)
    print(f"{name}: {round(Age_Group_Atrr,2)}%")

plt.figure(figsize=(10,6))
ax=sns.barplot(x=Attrition,y=Age_Group,hue=Age_Group,palette=["yellow","blue","orange","Green"],legend=True,width=0.5)
plt.title("Age Group vs Attrition Rate")
plt.ylabel("Age_Group",color="red")
plt.xlabel("Attrition %",color="red")
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f%%")
plt.xlim(0, max(Attrition) + 5)
plt.legend(loc="center left", bbox_to_anchor=(0.7,0.5))
plt.show()
# Overall attrition = 16%, Early Career = 35.77% which is very high comparing with Developing_Career
```

🟢 SECTION 3: Root Cause Analysis

Overtime vs Attrition 


```python
Over_Attrition=[]
Overtime=[]
for i in ibm_emp_att["OverTime"].unique():
    total=ibm_emp_att[(ibm_emp_att["OverTime"]==i)]["EmployeeNumber"].count()
    attrition=ibm_emp_att[(ibm_emp_att["Attrition"]=="Yes")&(ibm_emp_att["OverTime"]==i)]["EmployeeNumber"].count()
    overtime_attr=(attrition/total)*100
    Over_Attrition.append(overtime_attr)
    Overtime.append(i)
    print(f"OverTime of {i}: {round(overtime_attr,2)}%")

plt.figure(figsize=(10,6))    
x=sns.barplot(x=Overtime,y=Over_Attrition,hue=Overtime,palette="Set2",legend=True)
plt.title("Overtime vs Attrition",color="red")
plt.xlabel("OVERTIME",color="red")
plt.ylabel("Attrition%",color="red")
for container in x.containers:
    x.bar_label(container,fmt="%.2f%%")     
plt.show()
# Employees working overtime show significantly higher attrition (30.5%) compared to those without overtime (10.4%),
# suggesting workload pressure may be a key driver of employee turnover.
```

WorkLifeBalance vs Attrition

YearsAtCompany vs Attrition 


```python
Attrition=[]
Exp=[]

experience={"0–2_years":lambda x:x<=2,
"3–5_years":lambda x:(x>=3)&(x<=5),
"6–10_years":lambda x:(x>=6)&(x<=10), 
"10+_years":lambda x:x>10}

for labels,conditions in experience.items():
    group=ibm_emp_att[conditions(ibm_emp_att["YearsAtCompany"])]
    total=group["EmployeeNumber"].count()
    attrition=group[group["Attrition"]=="Yes"]["EmployeeNumber"].count()
    age_group_attrition=(attrition/total)*100
    Attrition.append(age_group_attrition)
    Exp.append(labels)
    print(f"{labels}: {round(age_group_attrition,2)}%")

plt.figure(figsize=(10,6))
x=sns.barplot(x=Exp,y=Attrition,hue=Exp,palette="Set2",legend=True)
plt.title("YearsAtCompany vs Attrition",color="red")
plt.xlabel("YearsAtCompany",color="red")
plt.ylabel("Attrition%",color="red")
plt.legend(title="YearsAtCompany")
for container in x.containers:
    x.bar_label(container,fmt="%.2f%%")
plt.ylim(0, max(Attrition) + 5)    
plt.show()      

# Employees within their first two years show significantly higher attrition (29.8%) compared to long-tenured employees (8.1%), 
# indicating that early-stage retention and onboarding experience may play a critical role in employee retention
```

🟢 SECTION 4: Combined Risk Patterns

1. Do employees in certain departments who work overtime show higher attrition?

```python
Dept=[]
Attrition=[]
for dept in ibm_emp_att["Department"].unique():
    group=ibm_emp_att[(ibm_emp_att["OverTime"]=="Yes")&(ibm_emp_att["Department"]==dept)]
    total=group["EmployeeNumber"].count()
    attrition=group[group["Attrition"]=="Yes"]["EmployeeNumber"].count()
    dept_overtime_attrition=(attrition/total)*100
    Dept.append(dept)
    Attrition.append(dept_overtime_attrition)
    print(f"{dept}: {round(dept_overtime_attrition,2)}%")
    
plt.figure(figsize=(10,6))
x=sns.barplot(x=Dept,y=Attrition,hue=Dept,palette="Set2",legend=True) 
plt.title("Department-wise Attrition Rate Among Overtime Employees",color="red")
plt.legend(title="Department")
plt.xlabel("Department",color="red")
plt.ylabel("Attrition",color="red")
for container in x.containers:
    x.bar_label(container,fmt="%.2f%%")
plt.ylim(0, max(Attrition) + 5)    
plt.show()    
```

2. Do early-career employees with overtime show higher attrition?

```python
Age_Group=[]
Attrition=[]

age_groups={
"Early_Career":lambda x:x<=25,
"Developing_Career":lambda x:(x>25)&(x<=35),
"Mid_Career":lambda x:(x>35)&(x<=45),
"Senior":lambda x:x>45
}

for ages,conditions in age_groups.items():
    group=ibm_emp_att[conditions(ibm_emp_att["Age"])&(ibm_emp_att["OverTime"]=="Yes")]
    total=group["EmployeeNumber"].count()
    attrition=group[group["Attrition"]=="Yes"]["EmployeeNumber"].count()
    overtime_ages=(attrition/total)*100
    Age_Group.append(ages)
    Attrition.append(overtime_ages)
    print(f"{ages}: {round(overtime_ages,2)}%")

plt.figure(figsize=(10,6))
x=sns.barplot(x=Age_Group,y=Attrition,hue=Age_Group,palette="Set2",legend=True)
plt.title("Impact of Overtime on Attrition by Career Stage")
plt.xlabel("Age_Group",color="red")
plt.ylabel("Attrition_%",color="red")
for container in x.containers:
    x.bar_label(container,fmt="%.2f%%")
plt.ylim(0,max(Attrition)+10)    
plt.show()

# Early-career employees working overtime exhibit extremely high attrition (64.1%), 
# indicating that workload pressure disproportionately affects younger employees who may already be exploring career opportunities.
```

3. Low Income + Overtime vs Attrition

```python
Income_Group=[]
Attrition=[]

income_group={
    "low_income":lambda x:x<=10000,
    "medium_income":lambda x:(x>10000)&(x<=15000),
    "high_income":lambda x:x>15000
}

for income,conditions in income_group.items():
    income_overtime=ibm_emp_att[conditions(ibm_emp_att["MonthlyIncome"])&(ibm_emp_att["OverTime"]=="Yes")]
    total=income_overtime["EmployeeNumber"].count()
    attrition=income_overtime[income_overtime["Attrition"]=="Yes"]["EmployeeNumber"].count()
    income_overtime_attrition=(attrition/total)*100
    Attrition.append(income_overtime_attrition)
    Income_Group.append(income)
    print(f"{income}: {round(income_overtime_attrition,2)}%")

plt.figure(figsize=(10,6))
x=sns.barplot(x=Income_Group,y=Attrition,hue=Income_Group,palette="Set2",legend=True)
plt.title("Impact of Low Income and Overtime on Employee Attrition")
plt.xlabel("Income_Group",color="red")
plt.ylabel("Attrition_%",color="red")
for container in x.containers:
    x.bar_label(container,fmt="%.2f%%")
plt.ylim(0,max(Attrition)+10)    
plt.show()    

# Low-income employees working overtime show significantly higher attrition (34.7%),
# suggesting that workload without adequate compensation may increase employee turnover risk.    
```

🟢 SECTION 5: Key Business Insights

1. The organization has an overall attrition rate of ~16%, indicating moderate employee turnover.

2. Sales department shows the highest attrition, suggesting higher workload or performance pressure in revenue-driven roles.

3. Employees working overtime exhibit significantly higher attrition (~30%), indicating workload pressure as a key driver.

4. Early-career employees (0–2 years) show the highest attrition (~29%), suggesting retention challenges during early tenure.

5. Employees reporting poor work-life balance show the highest attrition (~31%),highlighting the importance of work-life balance initiatives.

6. Low-income employees are more likely to leave compared to higher-income employees.

7. Combined risk analysis shows that early-career employees working overtime experience extremely high attrition (~64%), 
   representing the highest-risk employee segment.

Recommendations: 

• Reduce excessive overtime workload  
• Improve work-life balance programs  
• Strengthen early-career engagement and retention strategies  
• Review compensation structures for lower-income employees
